## Power Law Prior

$$\pi(\theta) = \theta^\alpha$$

```python
# This is normalized.
# alpha = spectral index

power_law_prior = bilby.core.prior.PowerLaw(minimum = 2, maximum = 7, alpha = -2) 

# The attributes can be accessed with the '.' operator.
```

```python
power_law_prior.minimum
```

```python
# This prior also has a number of methods.

print(power_law_prior.prob(val = 5))
print(power_law_prior.cdf(val = 5))
print(power_law_prior.sample(size = 1))
```

### Probability Density Function

```python
grid = np.linspace(2.0, 7.0, 100)
prob = power_law_prior.prob(grid)
plt.plot(grid, prob)
plt.xlabel("$\\theta$")
plt.ylabel("Probability Density")
plt.title("Prior Density")
plt.show()
```

```python
plt.plot(grid, prob)
plt.loglog()
plt.xlabel("$\\theta$")
plt.ylabel("Probability Density")
plt.title("Prior Density - LogLog")
plt.show()
```

### Cumulative Distribution Function

$$CDF(\theta) = \int_{\theta_{min}}^{\theta} \pi(\theta ') d\theta'$$

```python
cdf = power_law_prior.cdf(grid)
plt.plot(grid, cdf)
plt.xlabel("$\\theta$")
plt.ylabel("Cumulative Distribution")
plt.title("Cumulative Distribution Function")
plt.show()
```

### Sampling

```python
# Binning

samples = power_law_prior.sample(size = 100000)
plt.hist(samples, density = True, label = 'Binned Samples')   

# Analytic pdf

plt.plot(grid, prob, label = 'Analytic pdf')

plt.xlabel("$\\theta$")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
# Binning

samples = power_law_prior.sample(size = 100000)
plt.hist(samples, bins = 'fd', density = True, label = 'Binned Samples')   

# 'fd’ (Freedman Diaconis Estimator)
# Robust (resilient to outliers) estimator that takes into account data variability and data size.

# Analytic pdf

plt.plot(grid, prob, label = 'Analytic pdf')
plt.loglog()

plt.xlabel("$\\theta$")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

# Likelihoods in Bilby

Following classes are available:

- Likelihood
- ZeroLikelihood
- Analytical1DLikelihood
- GaussianLikelihood
- PoissonLikelihood
- ExponentialLikelihood
- StudentTLikelihood
- Multinomial
- AnalyticalMultidimensionalCovariantGaussian
- AnalyticalMultidimensionalBimodalCovariantGaussian
- JointLikelihood
- MarginalizedLikelihoodReconstructionError

Here we create our own likelihood:

$$\ln{\mathcal{L}} = -\frac{1}{2} \left(\sum_{i =1}^N {\left(\frac{{x_{i} - \mu}}{\sigma}\right)^2} + N\ln{(2\pi\sigma^2)}\right)$$

```python
class SimpleGaussianLikelihood(bilby.Likelihood):
    def __init__(self, data):
        super().__init__(parameters = {'mu': None, 'sigma': None})
        self.data = data
        self.N = len(data)
        
    def log_likelihood(self):
        mu = self.parameters['mu']
        sigma = self.parameters['sigma']
        diff = self.data - mu
        return -0.5 * (np.sum((diff/sigma)**2) + self.N * np.log(2 * np.pi * sigma**2))
```

```python
mu = 3
sigma = 4
data = np.random.normal(mu, sigma, 10000)

plt.hist(data, bins = 'fd')
plt.xlabel('Value')
plt.ylabel('Count')
plt.show()
```

```python
sns.distplot(data)    # normalized
```

```python
likelihood = SimpleGaussianLikelihood(data = data)
```

```python
priors = bilby.core.prior.PriorDict()
priors['mu'] = bilby.core.prior.Uniform(minimum = 0, maximum = 10, latex_label = "$\mu$")
priors['sigma'] = bilby.core.prior.Uniform(minimum = 0, maximum = 10, latex_label = "$\sigma$")
```

# Samplers in Bilby

- Default - dynesty
- emcee
- ptemcee
- nestle
- pymultinest

```python
result = bilby.run_sampler(likelihood = likelihood, priors = priors, outdir = 'MyLikelihood-run', 
                           label = 'MyLikelihood')
```

```python
result.plot_corner(truths=dict(mu=mu, sigma=sigma))
```
